```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import requests
import os
from random import randint
```

```python
def scroll(number_of_time):
    for i in range(number_of_time):
        driver.execute_script("window.scrollBy(0, 4200)")
        #setting sleep/waiting time for the page to load
        time.sleep(1)
```

```python
def image_search(keyword):
    global driver
    driver=webdriver.Chrome(r"D:\new_chrome_driver_may\chromedriver-win64\chromedriver.exe")
    website='https://images.google.com/'
    driver.get(website)
    driver.maximize_window()
    #Enter the elements name in to search bar
    search_element=driver.find_element(By.XPATH,value='//textarea[@class="gLFyf"]')
    search_element.send_keys(keyword)
    search_element.send_keys(Keys.ENTER)
    scroll(10)
```

```python
def downloading_img(keyword):
    #finding the number of images
    image_element=driver.find_elements(By.XPATH,value='//div[@class="F0uyec"]')
    #finding the src link and downloading it into local
    for i in range(len(image_element)):
        try:
            image_element[i].click()
            time.sleep(2)
            print('try')
            #getting the path of the image file from google
            img_url=driver.find_element(By.XPATH,value='//img[@class="sFlh5c pT0Scc iPVvYb"]').get_attribute('src')
            #defining the path of the local directory
            dir_path=r'C:\\Users\\asus\\Image Data Collection\\{}\\{}.jpg'.format(keyword,(randint(0, 1000000)))
            #creating the folder if not exist
            if not os.path.exists(r'C:\\Users\\asus\\Image Data Collection\\{}'.format(keyword)):
                os.makedirs(r'C:\\Users\\asus\\Image Data Collection\\{}'.format(keyword))
            save_image_from_url(img_url,dir_path)

        except:
            print('except')
            continue
```

```python
def save_image_from_url(url, save_path):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                f.write(response.content)
            print(f"Image saved successfully at {save_path}")
        else:
            print(f"Failed to download image from {url}. Status code: {response.status_code}")
    except Exception as e:
        print(f"Error occurred: {e}")
```
